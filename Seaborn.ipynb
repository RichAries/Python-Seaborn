# What is Seaborn? 


    Seaborn is a statistical plotting library. It has beautiful default styles. It also is designed to work very well with pandas dataframe objects.

## 1. Distribution Plot

```python
conda install seaborn
```

```python
import seaborn as sns
```

```python
#To see the visualiztion in the notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distplot show the only column distribution
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#the color of the hexagon represents the density.
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#To show the density of where these points match up the most
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#palette means change color
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

## 2. Categorical Plots

```python
#Sns.barplot x is categorical data, y is numerical data. 
#Check the command-press shift and tab.
sns.barplot(x='sex',y='total_bill', data=tips)
```

```python
#Calculate the standard deviation. estimator is aggregate function, like mean or sd, mean is default.
import numpy as np
import seaborn as sns
```

```python
sns.barplot(x='sex',y='total_bill', data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#Check another category
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#a scatter plot with category values
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#If there has overlapped points,use jitter=true to let them distribute evenly.
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
#swarmplot is a combination of violin plot and strip plot
#It doesn't fit well for large data
sns.swarmplot(x='day',y='total_bill',data=tips)
```

## 4.Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
#Customized pairplot what you want 
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

## 5. Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#markers means the shape of the point, and the scatter_kws means the size of markers
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex'
          ,aspect=0.6,size=10)
```

## 6. Style and Color

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#press the shift and tab to see what format can we use
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
#Another format called 'ticks'-had spine around it
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
#customized the spine. eg.right=True,top=Ture
sns.despine()
```

```python
#Adjusted the plot figure size
plt.figure(figsize =(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#Using system default format to adjust the plot size
sns.set_context('notebook',font_scale=0.9)
sns.countplot(x='sex',data=tips)
```
